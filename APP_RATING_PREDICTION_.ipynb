## Size column has sizes in Kb as well as Mb

###  Extract the numeric value from the column

```python
df.Size.value_counts()#getting  the no of apps per unique value in size column
```

```python
#Now Convert non nemuric value to nan value
df['Size'][df['Size' ] == 'Varies with devices'] = np.nan

#Finding NaN Values
df.Size.isna().sum()
```

```python
df.Size
```

### Multiply the value by 1,000, if size is mentioned in Mb

```python
#making function for converting all the kb and mb values to only kb and make it integer
def change_size(size):
    if 'M' in size:
        x = size[:-1]
        x = float(x)*1000
        return(x)
    else:
        return None
df["Size"] = df["Size"].map(change_size)

#filling Size which had NA 
df.Size.fillna(method = 'ffill', inplace = True)
```

```python
df.head(3)
```

```python
df['Reviews'] = df['Reviews'].astype(int)#Coneverting Reviews To Numeric Format
```

```python
df['Installs'] = df.Installs.str.replace('+', '')#removing all the + symbol that are available in installs to get it as integer
df['Installs'] = df.Installs.str.replace(',', '')#removing all the , symbol that are available in installs 
```

### LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn import linear_model, metrics

reg = linear_model.LinearRegression()

# train the model using the training sets
reg.fit(X_train, y_train)

# regression coefficients
print('Coefficients: ', reg.coef_)

# variance score: 1 means perfect prediction
print('Variance score: {}'.format(reg.score(X_test, y_test)))

# plot for residual error

## setting plot style
plt.style.use('fivethirtyeight')

## plotting residual errors in training data
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train,
			color = "green", s = 10, label = 'Train data')

## plotting residual errors in test data
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test,
			color = "blue", s = 10, label = 'Test data')

## plotting line for zero residual error
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)

## plotting legend
plt.legend(loc = 'upper right')

## plot title
plt.title("Residual errors")

## method call for showing the plot
plt.show()
```
